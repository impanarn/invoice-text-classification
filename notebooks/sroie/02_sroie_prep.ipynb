```python
import os
import pandas as pd

BASE_PATH = r"C:\Users\Impana\Downloads\invoice-classification\\"
D01_PATH = os.path.join(BASE_PATH, "data", "sroie", "D01_sroie_raw.csv")

df = pd.read_csv(D01_PATH)
print(df.shape, df['company'].nunique())
df.head(3)
```

```python
def categorize_company(name: str) -> str:
    n = str(name).upper()

    # FOOD / RESTAURANT / CAFE
    if any(k in n for k in ["RESTAURANT", "REST.", "CAFE", "FOOD", "KOPITIAM", "KITCHEN", "BAKERY"]):
        return "FOOD"

    # SUPERMARKET / CONVENIENCE / RETAIL
    if any(k in n for k in ["SUPERMARKET", "MART", "MINI MARKET", "MINIMARKET", "MINIMART",
                            "STORE", "MARKET", "SHOP", "DEPARTMENTAL"]):
        return "RETAIL"

    # STATIONERY / BOOK / PRINTING
    if any(k in n for k in ["BOOK", "STATIONERY", "PRINT", "PRESS", "PHOTO", "PHOTOSTAT"]):
        return "STATIONERY"

    # ELECTRONICS / IT
    if any(k in n for k in ["COMPUTER", "ELECTRIC", "ELECTRICAL", "ELECTRONIC", "IT ", "TECH"]):
        return "ELECTRONICS"

    # AUTO / WORKSHOP / TRANSPORT
    if any(k in n for k in ["WORKSHOP", "AUTO", "MOTOR", "TYRE", "TYRES", "CAR ", "MOTORS"]):
        return "AUTO"

    # CONSTRUCTION / HARDWARE
    if any(k in n for k in ["HARDWARE", "CONSTRUCTION", "CONTRACTOR", "ENGINEERING"]):
        return "HARDWARE"

    # HOTEL / LODGING / TRAVEL
    if any(k in n for k in ["HOTEL", "INN", "RESORT", "LODGE"]):
        return "HOTEL"

    # HEALTH / CLINIC / PHARMACY
    if any(k in n for k in ["CLINIC", "PHARMACY", "PHARMA", "HOSPITAL", "DENTAL"]):
        return "HEALTH"

    # SERVICES / GENERAL SERVICES
    if "SERVICES" in n or "SERVICE" in n:
        return "SERVICES"

    # Default
    return "OTHER"

df['category'] = df['company'].apply(categorize_company)
print(df['category'].value_counts())
```
